```python
rows_with_missing = df[df['Certificate'].isna()]
rows_with_missing
```

```python
df = df.dropna(subset=['Certificate'])
df
```

```python
# Group by decade and 'Certificate' and count occurrences (without creating a new column)
decade_counts = df.groupby([df['Released_Year'] // 10 * 10, 'Certificate']).size().reset_index(name='Count')

# Rename the decade column for clarity
decade_counts['Decade'] = decade_counts['Released_Year'].astype(str) + 's'

# Set up the plot
plt.figure(figsize=(10, 6))

# Plot the data
sns.lineplot(data=decade_counts, x='Decade', y='Count', hue='Certificate', marker='o')

# Add title and labels
plt.title('Certificate Distribution Over Decades')
plt.xlabel('Decade')
plt.ylabel('Count of Certificates')

# Show the plot
plt.grid(True)
plt.show()
```

```python
current_year = 2020

# Filter data for the last 5 years
df_past_5_years = df[df['Released_Year'] >= current_year - 5]

# Count the occurrences of each Certificate in the past 5 years
age_rating_counts = df_past_5_years['Certificate'].value_counts()

# Plot the pie chart for Certificate distribution in the past 5 years
plt.figure(figsize=(6, 6))
age_rating_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, legend=True)
plt.title(f'Certificate Distribution for the Past 5 Years ({current_year-5}-{current_year})')
plt.ylabel('')  # Hide the y-axis label
plt.show()
```

```python
df[df['Genre'].isna()]
```

```python
df['Genre'] = df['Genre'].str.split(',')
df['Genre']
```

```python
# Explode the 'Genre' column to have one genre per row
df_exploded = df.explode('Genre')

# Group by 'Certificate' and 'Genre', and count occurrences
genre_counts = df_exploded.groupby(['Certificate', 'Genre']).size().unstack(fill_value=0)

# Select top 5 genres for each Certificate
top_5_genres = genre_counts.apply(lambda x: x.nlargest(5), axis=1)

# Plot stacked column chart
ax = top_5_genres.plot(kind='bar', stacked=True, figsize=(10, 6))

# Set title and labels
plt.title('Top 5 Genres in Each Certificate')
plt.xlabel('Certificate')
plt.ylabel('Count of Top 5 Genres')

# Adjust the number of columns in the legend to split into two rows
ncol = len(top_5_genres.columns) // 3  # Divide number of genres by 2 to create two rows

# Adjust legend to show in two rows at the top
plt.legend(title='Genre', bbox_to_anchor=(0.5, 1.1), loc='center', ncol=ncol, frameon=False)

# Show the plot
plt.show()
```
